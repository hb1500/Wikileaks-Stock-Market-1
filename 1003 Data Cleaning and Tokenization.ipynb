#### Code below can be use to tokenize the words "content" and "Raw Content" in each  documents without changing the architecture of original data file

```python
def clean_tokenize(data):
    text = []
    text.append(data) 
    text = "".join(text)
        # split into words
    tokens = word_tokenize(text)
    tokens = [w.lower() for w in tokens]
    # remove all tokens that are not alphabetic
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]
    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()]
    # filter out stop words
    stop_words = set(stopwords.words('english'))
    words = [w for w in words if not w in stop_words]
    return words
```

```python
for key,value in config.items():
    config[key]['Content'] = clean_tokenize(config[key]['Content'])
    config[key]['Raw content'] = clean_tokenize(config[key]['Raw content'])
    print("file", key)
```
