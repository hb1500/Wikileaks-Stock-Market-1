```python
from datetime import datetime
def tranf_date(x):
    return datetime.strptime(x, '%m/%d/%Y').strftime('%Y%m%d') 

#### formating date column
```

```python
df['date'] = df['date'].apply(lambda i:  tranf_date(i))
```

### create vector for exchange rate: 

```python
def tovector(data, column, row):
    return np.array(data[column][row-15:row])    
```

```python
df_ex = pd.DataFrame()
```

```python
for column in ['australia', 'brazil', 'canada', 'china', 'denmark','hong kong', 'india', 'japan', 'korea', 'malaysia', 'mexico','new zealand', 'norway', 'sweden', 'south africa', 'singapore','sri lanka', 'switzerland', 'taiwan', 'thailand', 'united kingdom','venezuela']:
    # standardize the column
    df[column] = df[column].apply(lambda x: (x - df[column].mean())/df[column].std())
    l = []
    name = column + str('_vec')
    for row in range(15, len(df[column])+1):    
        l.append(tovector(df, column, row)) 
    df_ex[name] =pd.DataFrame({name: l})
```

### create vector for log_diff exchange rate:

```python
for column in ['australia_logdff', 'brazil_logdff', 'canada_logdff','china_logdff', 'denmark_logdff', 'hong kong_logdff', 'india_logdff','japan_logdff', 'korea_logdff', 'malaysia_logdff', 'mexico_logdff','new zealand_logdff', 'norway_logdff', 'sweden_logdff','south africa_logdff', 'singapore_logdff', 'sri lanka_logdff','switzerland_logdff', 'taiwan_logdff', 'thailand_logdff','united kingdom_logdff', 'venezuela_logdff']:
    l = []
    name = column + str('_vec')
    for row in range(15, len(df[column])+1):    
        l.append(tovector(df, column, row)) 
    df_ex[name] =pd.DataFrame({name: l})
```

```python
def writeToJSONFile(path, fileName, data):
    '''
    merge different dictionaries from different files
    input:
        path: path to store the output
        fileName: the stored file name
        data: list of dictionaries
    output:
        merged dictionaries(dataframe, if using pandas read)
    '''

    if not os.path.exists(path):
        '''
        if there is no output directory, creating one
        '''
        os.makedirs(path)
    filePathNameWExt = './' + path + '/' + fileName
    with open(filePathNameWExt, 'w') as fp:
        data.to_json(fp)
```
